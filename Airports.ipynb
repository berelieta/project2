```python
import pandas as pd
```

```python
flights_csv = "DelayedFlights.csv"
```

```python
flights = pd.read_csv(flights_csv)
```

```python
flights.head()
```

```python
flights_df = flights[["Origin","Dest", "ArrDelay", "DepDelay"]]
```

```python
airports_csv = "airports.csv"
airports = pd.read_csv(airports_csv, encoding='latin1')
airports.head()
```

```python
airports_df = airports[["AIRPORT", "IATA_CODE", "STATE"]]
airports_df
```

```python
airports_df= airports_df.rename(columns = {"IATA_CODE":"Origin"})
airports_df
```

```python
flights_df.head(3)
```

```python
complete = pd.merge(flights_df, airports_df, how="left", on=["Origin"])
complete.head(5)
```

```python
flights_origin = complete.groupby(["Origin", "AIRPORT", "STATE"])["ArrDelay"].mean()
flights_origin.head()
```

```python
worst_airports = flights_origin.sort_values(ascending=False).head(10)
worst_airports
```

```python
best_airports = flights_origin.sort_values(ascending=True).head(10)
best_airports
```

```python
airports_info = pd.concat([best_airports, worst_airports], axis=0)
```

```python
airports_info = airports_info.sort_values(ascending = True)
type(airports_info)
airports_info = airports_info.to_frame().reset_index("AIRPORT")
airports_info = airports_info.rename(columns= {"ArrDelay": 'value', "AIRPORT": "x"})
```

```python
airports_info
```

```python
airports_info.to_json(orient="table")
```
